```python
import gseapy as gp
from gseapy import Biomart
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
```

```python
bm = Biomart()
marts = bm.get_marts()
print(marts)
datasets = bm.get_datasets(mart='ENSEMBL_MART_ENSEMBL')
```

```python
attrs = bm.get_attributes(dataset='hsapiens_gene_ensembl')
attrs
```

```python
gene_names = pd.read_csv("genes.tsv", sep = '\t')
len(gene_names['gene'].values.astype(str).tolist())
```

```python
queries ={'external_gene_name': gene_names['gene'].values.astype(str).tolist()}
results = bm.query(dataset='hsapiens_gene_ensembl',
                   attributes=['external_gene_name', 'entrezgene_id'],
                   filters=queries)
results.to_csv('enterz_id.tsv', sep = '\t', index = True)
```
